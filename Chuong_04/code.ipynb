```python
import pandas as pd
import numpy as np

# --- TẠO DỮ LIỆU MẪU CHO EURO 2012 ---
euro12_data = {
    'Team': ['Italy', 'England', 'Russia', 'Germany', 'Greece', 'Spain', 'Portugal', 'France'],
    'Goals': [6, 5, 5, 10, 5, 12, 6, 3],
    'Shooting Accuracy': ['50.0%', '48.3%', '45.0%', '55.9%', '40.2%', '56.0%', '49.5%', '43.1%'],
    'Yellow Cards': [16, 9, 6, 4, 9, 11, 12, 6],
    'Red Cards': [0, 0, 1, 0, 1, 0, 0, 0],
    'Subs on': [18, 14, 10, 15, 12, 16, 14, 15],
    'Corners Taken': [30, 25, 20, 35, 15, 40, 32, 22]
}
pd.DataFrame(euro12_data).to_csv('euro12.csv', index=False)

# --- GIẢI BÀI 4.20 ---
# 1. Đọc dữ liệu và in thông tin cơ bản
euro12 = pd.read_csv('euro12.csv')
print(f"Type: {type(euro12)}")
print(f"Shape: {euro12.shape}")
print(f"Columns: {euro12.columns.tolist()}")

# 1 (Tiếp). In giá trị cột Goals
print("\n--- Cột Goals ---")
print(euro12['Goals'])

# 2. Số đội tham gia
print(f"\nSố đội tham gia: {euro12['Team'].nunique()}")

# 3. In thông tin
print("\n--- Thông tin Euro2012 ---")
euro12.info()

# 4. Tạo dataframe discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]

# 5. Sắp xếp discipline giảm dần
discipline_sorted = discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending=False)
print("\n--- Discipline Sorted ---")
display(discipline_sorted)

# 6. a) Trung bình Yellow Cards; b) Đội > 6 bàn thắng
print(f"\nTrung bình Yellow Cards: {euro12['Yellow Cards'].mean()}")
teams_over_6 = euro12[euro12['Goals'] > 6]
print("\nĐội ghi hơn 6 bàn:")
print(teams_over_6['Team'].tolist())

# 7. Đội bắt đầu bằng 'G'
teams_G = euro12[euro12['Team'].str.startswith('G')]
print("\nĐội bắt đầu bằng 'G':", teams_G['Team'].tolist())

# 8. 7 cột đầu
print("\n--- 7 cột đầu ---")
display(euro12.iloc[:, :7])

# 9. Tất cả cột trừ 3 cột cuối
print("\n--- Bỏ 3 cột cuối ---")
display(euro12.iloc[:, :-3])

# 10. In các cột cụ thể
print("\n--- Cột Team, Goals, Accuracy, Cards ---")
display(euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']])

# 11. Lọc England, Italy, Russia (chỉ Team và Accuracy)
filtered_teams = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]
display(filtered_teams)
```

```python
import pandas as pd

# --- TẠO DỮ LIỆU MẪU CHO DRINKS ---
drinks_data = {
    'country': ['Vietnam', 'USA', 'France', 'Nigeria', 'Japan', 'Brazil', 'Germany', 'Australia'],
    'beer_servings': [300, 249, 127, 31, 77, 245, 346, 261],
    'spirit_servings': [100, 158, 151, 2, 202, 145, 117, 72],
    'wine_servings': [1, 84, 370, 1, 16, 16, 175, 212],
    'total_litres_of_pure_alcohol': [4.5, 8.7, 11.8, 0.5, 2.0, 7.2, 11.3, 10.4],
    'continent': ['AS', 'NA', 'EU', 'AF', 'AS', 'SA', 'EU', 'OC']
}
pd.DataFrame(drinks_data).to_csv('drinks.csv', index=False)

# --- GIẢI BÀI 4.21 ---
# 1. Đọc dữ liệu (index_col là cột đầu tiên 'country')
drink = pd.read_csv('drinks.csv', index_col=0)
print(f"Type: {type(drink)}, Shape: {drink.shape}")
print(f"Columns: {drink.columns.tolist()}")
print("\n--- 5 dòng đầu ---")
display(drink.head(5))

# 2. Số lượng bia trung bình mỗi châu lục
print("\n--- Bia trung bình theo Châu lục ---")
print(drink.groupby('continent')['beer_servings'].mean())

# 3. Thống kê wine theo châu lục
print("\n--- Thống kê Wine theo Châu lục ---")
display(drink.groupby('continent')['wine_servings'].describe())

# 4. Trung bình tất cả các loại bia/rượu theo châu lục
print("\n--- Mean bia/rượu theo Châu lục ---")
display(drink.groupby('continent').mean(numeric_only=True))

# 5. Trung vị (Median) theo châu lục
print("\n--- Median bia/rượu theo Châu lục ---")
display(drink.groupby('continent').median(numeric_only=True))

# 6. Spirit (Rượu mạnh): Mean, Max, Min theo châu lục
print("\n--- Spirit stats theo Châu lục ---")
display(drink.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min']))

# 7. Sắp xếp theo lượng bia tiêu thụ
drink_sorted = drink.sort_values('beer_servings')
print("\n--- 5 quốc gia tiêu thụ bia NHIỀU nhất ---")
display(drink_sorted.tail(5))
print("\n--- 5 quốc gia tiêu thụ bia ÍT nhất ---")
display(drink_sorted.head(5))
```

```python
import pandas as pd
import numpy as np
import os

# --- BƯỚC TỰ TẠO DATA MẪU ---
if not os.path.exists('stocks1.csv'):
    stocks1_data = {
        'date': ['2023-01-01', '2023-01-01', '2023-01-02'],
        'symbol': ['AAPL', 'GOOGL', 'AAPL'],
        'open': [150.0, 2800.0, 155.0],
        'high': [160.0, np.nan, 158.0], # Có giá trị Null để làm ý 2
        'low': [148.0, 2780.0, np.nan],  # Có giá trị Null để làm ý 2
        'close': [158.0, 2810.0, 156.0],
        'volume': [1000000, 500000, 1100000]
    }
    pd.DataFrame(stocks1_data).to_csv('stocks1.csv', index=False)

if not os.path.exists('stocks2.csv'):
    stocks2_data = {
        'date': ['2023-01-03', '2023-01-03'],
        'symbol': ['AAPL', 'GOOGL'],
        'open': [156.0, 2815.0],
        'high': [162.0, 2830.0],
        'low': [155.0, 2805.0],
        'close': [160.0, 2825.0],
        'volume': [1200000, 600000]
    }
    pd.DataFrame(stocks2_data).to_csv('stocks2.csv', index=False)

if not os.path.exists('companies.csv'):
    comp_data = {
        'name': ['AAPL', 'GOOGL', 'MSFT'],
        'employees': [150000, 140000, 180000],
        'headquarters_city': ['Cupertino', 'Mountain View', 'Redmond'],
        'headquarters_state': ['CA', 'CA', 'WA']
    }
    pd.DataFrame(comp_data).to_csv('companies.csv', index=False)

# --- BẮT ĐẦU GIẢI BÀI 4.22 ---
# 1. Đọc dữ liệu
stocks1 = pd.read_csv('stocks1.csv')
stocks2 = pd.read_csv('stocks2.csv')
companies = pd.read_csv('companies.csv')

# 2. Xử lý Null theo quy tắc đề bài
stocks1['high'] = stocks1.groupby('symbol')['high'].transform(lambda x: x.fillna(x.max()))
stocks1['low'] = stocks1.groupby('symbol')['low'].transform(lambda x: x.fillna(x.min()))

# 3. Gộp stocks1 và stocks2
stocks = pd.concat([stocks1, stocks2], axis=0, ignore_index=True)

# 4. Gộp stocks và companies
stocks_companies = pd.merge(stocks, companies, left_on='symbol', right_on='name')

# 5 & 6. Thống kê
avg_per_company = stocks_companies.groupby('name')[['open', 'high', 'low', 'close', 'volume']].mean()
close_stats = stocks_companies.groupby('name')['close'].agg(['mean', 'max', 'min'])

# 7. Cột parsed_time
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])

# 8. Cột result
stocks_companies['result'] = np.where(stocks_companies['close'] > stocks_companies['open'], 'up', 'down')

print("--- Kết quả Bài 4.22 (5 dòng đầu) ---")
display(stocks_companies.head())
```

```python
# --- TẠO DATA MẪU CHO MOVIELENS ---
movies_demo = {
    'movieId': [1, 2, 3],
    'title': ['Toy Story (1995)', 'Jumanji (1995)', 'Grumpier Old Men (1995)'],
    'genres': ['Adventure|Animation', 'Adventure|Children', 'Comedy|Romance']
}
ratings_demo = {
    'userId': [1, 1, 2, 2],
    'movieId': [1, 2, 1, 3],
    'rating': [4.0, 3.5, 5.0, 2.0],
    'timestamp': [964982703, 964981247, 964984002, 964984050]
}
pd.DataFrame(movies_demo).to_csv('movies.csv', index=False)
pd.DataFrame(ratings_demo).to_csv('ratings.csv', index=False)

# --- GIẢI BÀI 4.23 ---
# Đọc và Xử lý
df_movies = pd.read_csv('movies.csv')
df_ratings = pd.read_csv('ratings.csv')

# Gộp và Thống kê
movie_combined = pd.merge(df_movies, df_ratings, on='movieId')
movie_combined['date'] = pd.to_datetime(movie_combined['timestamp'], unit='s')

# Tính rating trung bình mỗi phim
top_movies = movie_combined.groupby('title')['rating'].mean().sort_values(ascending=False)

print("--- Kết quả Bài 4.23 (Rating trung bình) ---")
print(top_movies)
display(movie_combined.head())
```
