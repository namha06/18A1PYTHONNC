```python
import threading
import time

def print_thread_name():
    print(f"Luồng đang chạy: {threading.current_thread().name}")

# Tạo danh sách các luồng
threads = []
for i in range(5):
    t = threading.Thread(target=print_thread_name, name=f"Luồng-{i+1}")
    threads.append(t)
    t.start()

# Chờ các luồng hoàn thành
for t in threads:
    t.join()
```

```python
import threading
import time

def download_file(file_name):
    print(f"Bắt đầu tải: {file_name}...")
    time.sleep(2)  # Mô phỏng thời gian tải tệp
    print(f"Tải xong: {file_name}")

files = ["file1.zip", "file2.jpg", "file3.mp4"]
threads = []

for file in files:
    t = threading.Thread(target=download_file, args=(file,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
import threading

def find_even(start, end):
    evens = [n for n in range(start, end + 1) if n % 2 == 0]
    print(f"Số chẵn: {evens}")

def find_odd(start, end):
    odds = [n for n in range(start, end + 1) if n % 2 != 0]
    print(f"Số lẻ: {odds}")

t1 = threading.Thread(target=find_even, args=(30, 50))
t2 = threading.Thread(target=find_odd, args=(30, 50))

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
import threading

def factorial(n):
    res = 1
    for i in range(1, n + 1):
        res *= i
    print(f"Giai thừa của {n} là: {res}")

numbers = [5, 7, 10]
for num in numbers:
    threading.Thread(target=factorial, args=(num,)).start()
```

```python
import threading
import requests

def fetch_url(url):
    try:
        response = requests.get(url)
        print(f"URL: {url} - Trạng thái: {response.status_code}")
    except Exception as e:
        print(f"Lỗi khi truy cập {url}: {e}")

urls = ["https://google.com", "https://python.org", "https://github.com"]
for url in urls:
    threading.Thread(target=fetch_url, args=(url,)).start()
```

```python
import threading

def print_numbers(is_even, threshold):
    label = "Chẵn" if is_even else "Lẻ"
    start = 0 if is_even else 1
    nums = [i for i in range(start, threshold + 1, 2)]
    print(f"Các số {label} tăng dần đến {threshold}: {nums}")

threshold = 20
t_even = threading.Thread(target=print_numbers, args=(True, threshold))
t_odd = threading.Thread(target=print_numbers, args=(False, threshold))

t_even.start()
t_odd.start()

t_even.join()
t_odd.join()
```

```python
import threading
import time

def print_time(thread_name, delay, iterations):
    print(f"Starting {thread_name}")
    for i in range(iterations):
        time.sleep(delay)
        # Lấy thời gian hiện tại định dạng theo mẫu trong ảnh
        current_time = time.strftime("%b %d %H:%M:%S %Y", time.localtime())
        print(f"{thread_name}: Web {current_time}")
    print(f"Exiting {thread_name}")

# Khởi tạo luồng chính
print("Starting Main Thread")

# Tạo các luồng con với độ trễ và số lần lặp khác nhau để mô phỏng kết quả như ảnh
t1 = threading.Thread(target=print_time, args=("Google", 1, 4))
t2 = threading.Thread(target=print_time, args=("Yahoo", 2, 3))
t3 = threading.Thread(target=print_time, args=("Facebook", 3, 2))

# Chạy các luồng
t1.start()
t2.start()
t3.start()

# Chờ các luồng con hoàn thành
t1.join()
t2.join()
t3.join()

print("Exiting Main Thread")
```

```python
import threading
import random

# ====== Tạo list ngẫu nhiên ======
n = 40
lst = [random.randint(0, 10) for _ in range(n)]
print("List:", lst)
print("Số phần tử:", len(lst))

# ====== Hàm cho Thread ======
def sum_sublist(sublist, results, index):
    s = sum(sublist)
    results[index] = s
    print(f"Thread {index + 1}: {sublist} -> Tổng = {s}")

# ====== Chia list & tạo Thread ======
num_threads = 4
chunk_size = len(lst) // num_threads

threads = []
results = [0] * num_threads

for i in range(num_threads):
    start = i * chunk_size
    end = start + chunk_size
    sublist = lst[start:end]

    t = threading.Thread(
        target=sum_sublist,
        args=(sublist, results, i)
    )
    threads.append(t)
    t.start()

# ====== Chờ các Thread hoàn thành ======
for t in threads:
    t.join()

# ====== Kết quả cuối ======
print("\nTổng các list con:", results)
print("Tổng toàn bộ list:", sum(results))
```

```python
import threading
import time

# Hàm thực thi cho từng luồng
def web_service(name, delay, iterations):
    print(f"Starting {name}")
    for i in range(iterations):
        # Tạm dừng luồng theo thời gian delay
        time.sleep(delay)
        # Lấy thời gian hiện tại định dạng: Tháng Ngày Giờ:Phút:Giây Năm
        timestamp = time.strftime("%b %d %H:%M:%S %Y", time.localtime())
        print(f"{name}: Web {timestamp}")
    print(f"Exiting {name}")

# Khởi tạo luồng chính
print("Starting Main Thread")

# Tạo 3 luồng con với các thông số khác nhau để mô phỏng thứ tự kết thúc
# Google lặp 4 lần, mỗi lần cách nhau 1s
t1 = threading.Thread(target=web_service, args=("Google", 1, 4))
# Yahoo lặp 3 lần, mỗi lần cách nhau 2s
t2 = threading.Thread(target=web_service, args=("Yahoo", 2, 3))
# Facebook lặp 2 lần, mỗi lần cách nhau 3s
t3 = threading.Thread(target=web_service, args=("Facebook", 3, 2))

# Bắt đầu thực thi các luồng
t1.start()
t2.start()
t3.start()

# Chờ tất cả các luồng con kết thúc trước khi kết thúc luồng chính
t1.join()
t2.join()
t3.join()

print("Exiting Main Thread")
```

```python
import threading
import random

def find_max_sublist(thread_id, sub_list, results):
    if not sub_list:
        return
    max_val = max(sub_list)
    list_str = "; ".join(map(str, sub_list))
    print(f"Thread {thread_id} xử lý list: [{list_str}]")
    print(f"-> Max của Thread {thread_id} là: {max_val}")
    results[thread_id] = max_val

# --- BƯỚC NHẬP DỮ LIỆU CÓ KIỂM TRA LỖI ---
while True:
    try:
        n_input = input("Nhập số phần tử n: ")
        if n_input.strip() == "": # Kiểm tra nếu để trống
            print("Vui lòng không để trống!")
            continue
        n = int(n_input)
        break # Nếu nhập đúng số thì thoát vòng lặp
    except ValueError:
        print("Lỗi: Bạn phải nhập vào một số nguyên (ví dụ: 10, 20, 100)!")

main_list = [random.randint(0, 100) for _ in range(n)]
print(f"List tổng quát: {main_list}")

while True:
    try:
        num_threads_input = input("Nhập số lượng luồng: ")
        num_threads = int(num_threads_input)
        if num_threads <= 0:
            print("Số luồng phải lớn hơn 0!")
            continue
        break
    except ValueError:
        print("Lỗi: Vui lòng nhập số nguyên cho số lượng luồng!")

# --- PHẦN CHIA LUỒNG (GIỮ NGUYÊN) ---
threads = []
thread_results = {}
chunk_size = max(1, n // num_threads)

for i in range(num_threads):
    start = i * chunk_size
    sub_list = main_list[start:] if i == num_threads - 1 else main_list[start : start + chunk_size]
    
    if sub_list:
        t = threading.Thread(target=find_max_sublist, args=(i + 1, sub_list, thread_results))
        threads.append(t)
        t.start()

for t in threads:
    t.join()

if thread_results:
    final_max = max(thread_results.values())
    print("-" * 30)
    print(f"Giá trị lớn nhất trong toàn bộ list là: {final_max}")
```
