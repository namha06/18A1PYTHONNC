```python
import numpy as np
import os

# Đọc dữ liệu từ file baseball_2D.txt (nếu có), nếu không thì dùng dữ liệu mẫu
if os.path.exists("baseball_2D.txt"):
    baseball = np.loadtxt("baseball_2D.txt").tolist()
else:
    baseball = [
        [180, 80],
        [175, 75],
        [190, 90],
        [185, 85],
        [178, 78]
    ]

# Tạo 2D numpy array np_baseball, kiểm tra kiểu dữ liệu và kích thước
np_baseball = np.array(baseball)
print("dtype:", np_baseball.dtype)
print("shape:", np_baseball.shape)

# In dòng thứ 50
if np_baseball.shape[0] >= 50:
    print("Dòng 50:", np_baseball[49])
else:
    print("Không đủ 50 dòng")

# Tạo numpy array np_weight từ cột cân nặng
np_weight = np_baseball[:, 1]
print("np_weight:", np_weight)

# Chiều cao và cân nặng của vận động viên thứ 124
if np_baseball.shape[0] >= 124:
    print("VĐV thứ 124:", np_baseball[123])
else:
    print("Không đủ 124 vận động viên")

# Chiều cao trung bình và cân nặng trung bình
print("Chiều cao TB:", np.mean(np_baseball[:, 0]))
print("Cân nặng TB:", np.mean(np_baseball[:, 1]))

# Mối tương quan chiều cao – cân nặng
print("Ma trận tương quan:")
print(np.corrcoef(np_baseball[:, 0], np_baseball[:, 1]))
```

```python
# import thư viện
import numpy as np
import os

# đọc dữ liệu từ file, nếu không có thì dùng dữ liệu mẫu
if os.path.exists("positions.txt") and os.path.exists("heights.txt"):
    positions = np.loadtxt("positions.txt", dtype=str)
    heights = np.loadtxt("heights.txt")
else:
    positions = np.array(['GK','M','A','D','GK','M','D','A','GK','M'])
    heights = np.array([190, 175, 180, 185, 192, 178, 182, 176, 195, 170])

# Ý 1: tạo numpy array np_positions và kiểm tra kiểu dữ liệu
np_positions = np.array(positions)
print("np_positions dtype:", np_positions.dtype)

# Ý 2: tạo numpy array np_heights và kiểm tra kiểu dữ liệu
np_heights = np.array(heights)
print("np_heights dtype:", np_heights.dtype)

# Ý 3: tính chiều cao trung bình của các GK
print("Chiều cao TB GK:", np.mean(np_heights[np_positions == 'GK']))

# Ý 4: tính chiều cao trung bình của các vị trí khác GK
print("Chiều cao TB vị trí khác GK:", np.mean(np_heights[np_positions != 'GK']))

# Ý 5: tìm cầu thủ thấp nhất và cao nhất
idx_min = np.argmin(np_heights)
idx_max = np.argmax(np_heights)
print("Thấp nhất:", np_positions[idx_min], np_heights[idx_min])
print("Cao nhất:", np_positions[idx_max], np_heights[idx_max])
```

```python
import numpy as np
from io import StringIO

# ===== DỮ LIỆU EURO 2012 
csv_data = """Team,Goals,Shooting Accuracy,Yellow Cards,Red Cards
Croatia,4,51.9,9,0
Czech Republic,4,41.3,7,0
Denmark,4,38.5,4,0
England,5,50.0,5,0
France,3,37.9,6,0
Germany,10,56.3,4,0
Greece,5,32.1,9,1
Italy,6,43.0,16,0
Netherlands,2,44.1,5,0
Poland,2,31.7,7,1
Portugal,6,41.5,12,0
Republic of Ireland,1,36.8,6,1
Russia,5,39.9,6,0
Spain,12,55.9,11,0
Sweden,5,47.2,7,0
Ukraine,2,21.2,5,0
"""

# ===== ĐỌC DỮ LIỆU =====
euro12 = np.genfromtxt(
    StringIO(csv_data),
    delimiter=",",
    names=True,
    dtype=None,
    encoding="utf-8"
)

# 1. In giá trị cột Goals
print("1. Goals:")
print(euro12["Goals"])

# 2. Có bao nhiêu đội tham gia Euro2012
print("\n2. Số đội tham gia:")
print(len(euro12))

# 3. Thông tin euro12
print("\n3. Thông tin euro12:")
print("Type:", type(euro12))
print("Shape:", euro12.shape)
print("Columns:", euro12.dtype.names)

# 4. Tạo euro12_discipline
euro12_discipline = euro12[["Team", "Yellow_Cards", "Red_Cards"]]
print("\n4. euro12_discipline:")
print(euro12_discipline)

# 5. Sắp xếp giảm dần theo Red Cards, Yellow Cards (ĐÚNG)
idx = np.lexsort((
    -euro12_discipline["Yellow_Cards"],  # khóa phụ
    -euro12_discipline["Red_Cards"]      # khóa chính
))
print("\n5. Sắp xếp discipline giảm dần:")
print(euro12_discipline[idx])

# 6a. Tính trung bình Yellow Cards (ĐÚNG)
print("\n6a. Trung bình Yellow Cards:")
print(np.mean(euro12["Yellow_Cards"]))

# 6b. Lọc các đội ghi hơn 6 bàn thắng (ĐÚNG)
print("\n6b. Các đội ghi hơn 6 bàn:")
print(euro12[euro12["Goals"] > 6]["Team"])

# 7. In các đội tên bắt đầu bằng 'G'
print("\n7. Đội bắt đầu bằng 'G':")
print(euro12[np.char.startswith(euro12["Team"], "G")]["Team"])

# 8. In 7 cột đầu (dataset này có 5 cột)
print("\n8. Các cột hiện có:")
print(euro12)

# 9. In tất cả các cột trừ 3 cột cuối
print("\n9. Tất cả cột trừ 3 cột cuối:")
print(euro12[["Team", "Goals"]])

# 10. In các cột Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards
print("\n10. Các cột yêu cầu:")
print(euro12[["Team", "Goals", "Shooting_Accuracy", "Yellow_Cards", "Red_Cards"]])

# 11. Team & Shooting Accuracy của England, Italy, Russia
print("\n11. England, Italy, Russia:")
mask = np.isin(euro12["Team"], ["England", "Italy", "Russia"])
print(euro12[mask][["Team", "Shooting_Accuracy"]])
```

```python
import numpy as np
from io import StringIO

# ===== DỮ LIỆU drinks.csv 
csv_data = """country,beer_servings,spirit_servings,wine_servings,total_litres_of_pure_alcohol,continent
Afghanistan,0,0,0,0.0,Asia
Albania,89,132,54,4.9,Europe
Algeria,25,0,14,0.7,Africa
Andorra,245,138,312,12.4,Europe
Angola,217,57,45,5.9,Africa
Argentina,193,25,221,8.3,South America
Australia,261,72,212,10.4,Oceania
Austria,279,75,191,9.7,Europe
Bangladesh,0,0,0,0.0,Asia
Belgium,295,84,212,10.5,Europe
Brazil,245,145,16,7.2,South America
Canada,240,122,100,8.2,North America
China,79,192,8,5.0,Asia
Denmark,224,81,278,10.4,Europe
Egypt,6,4,1,0.1,Africa
Finland,263,133,97,10.0,Europe
France,127,151,370,11.8,Europe
Germany,346,117,175,11.3,Europe
Greece,133,112,218,8.3,Europe
India,9,114,0,2.2,Asia
Italy,85,42,237,6.5,Europe
Japan,77,202,16,7.0,Asia
Mexico,238,68,5,5.5,North America
Norway,169,71,129,6.7,Europe
Portugal,194,67,339,11.0,Europe
Russia,247,326,73,11.5,Europe
Spain,284,157,112,10.0,Europe
USA,249,158,84,8.7,North America
Vietnam,111,2,1,2.0,Asia
"""

# ===== ĐỌC DỮ LIỆU =====
drink = np.genfromtxt(
    StringIO(csv_data),
    delimiter=",",
    names=True,
    dtype=None,
    encoding="utf-8"
)

# 1. Thông tin dữ liệu
print("1. Thông tin drink:")
print("Type:", type(drink))
print("Shape:", drink.shape)
print("Columns:", drink.dtype.names)
print("\nHead (5 dòng đầu):")
print(drink[:5])
print("\nTail (5 dòng cuối):")
print(drink[-5:])

# ===== GROUP BY THEO CONTINENT =====
continents = np.unique(drink["continent"])

# 2. Số lượng bia tiêu thụ trung bình mỗi châu lục
print("\n2. Beer servings trung bình mỗi châu lục:")
for c in continents:
    print(c, ":", np.mean(drink["beer_servings"][drink["continent"] == c]))

# 3. Describe wine_servings theo châu lục
print("\n3. Thống kê wine_servings theo châu lục:")
for c in continents:
    data = drink["wine_servings"][drink["continent"] == c]
    print(c, ": min =", np.min(data),
          "max =", np.max(data),
          "mean =", np.mean(data),
          "std =", np.std(data))

# 4. Mean bia & rượu theo châu lục
print("\n4. Mean beer & wine theo châu lục:")
for c in continents:
    mask = drink["continent"] == c
    print(c,
          "Beer =", np.mean(drink["beer_servings"][mask]),
          "Wine =", np.mean(drink["wine_servings"][mask]))

# 5. Median bia & rượu theo châu lục
print("\n5. Median beer & wine theo châu lục:")
for c in continents:
    mask = drink["continent"] == c
    print(c,
          "Beer =", np.median(drink["beer_servings"][mask]),
          "Wine =", np.median(drink["wine_servings"][mask]))

# 6. Spirit_servings: mean, max, min mỗi châu lục
print("\n6. Spirit_servings theo châu lục:")
for c in continents:
    data = drink["spirit_servings"][drink["continent"] == c]
    print(c,
          "Mean =", np.mean(data),
          "Max =", np.max(data),
          "Min =", np.min(data))

# 7. Sắp xếp theo beer_servings tăng dần
sorted_drink = np.sort(drink, order="beer_servings")

print("\n7a. 5 quốc gia uống bia nhiều nhất:")
print(sorted_drink[-5:][["country", "beer_servings"]])

print("\n7b. 5 quốc gia uống bia ít nhất:")
print(sorted_drink[:5][["country", "beer_servings"]])
```

```python
import pandas as pd
import numpy as np
from io import StringIO

# =========================
# DỮ LIỆU stocks1.csv
# =========================
stocks1_csv = """date,symbol,open,high,low,close,volume
2023-01-01,AAPL,130,135,128,134,10000
2023-01-02,AAPL,134,,132,133,12000
2023-01-01,MSFT,220,225,218,224,15000
2023-01-02,MSFT,224,226,,225,14000
"""

stocks1 = pd.read_csv(StringIO(stocks1_csv))

print("1a. stocks1 head:")
print(stocks1.head())
print("\n1a. stocks1 tail:")
print(stocks1.tail())
print("\nDtypes stocks1:")
print(stocks1.dtypes)
print("\nInfo stocks1:")
stocks1.info()

# =========================
# DỮ LIỆU stocks2.csv
# =========================
stocks2_csv = """date,symbol,open,high,low,close,volume
2023-01-03,AAPL,133,136,131,135,11000
2023-01-03,MSFT,225,228,223,227,16000
"""

stocks2 = pd.read_csv(StringIO(stocks2_csv))

print("\n1b. stocks2 head:")
print(stocks2.head())
print("\n1b. stocks2 tail:")
print(stocks2.tail())
print("\nDtypes stocks2:")
print(stocks2.dtypes)
print("\nInfo stocks2:")
stocks2.info()

# =========================
# DỮ LIỆU companies.csv
# =========================
companies_csv = """symbol,name,employees,headquarters_city,headquarters_state
AAPL,Apple,160000,Cupertino,CA
MSFT,Microsoft,221000,Redmond,WA
"""

companies = pd.read_csv(StringIO(companies_csv))

print("\n1c. companies:")
print(companies.head())
print("\nDtypes companies:")
print(companies.dtypes)
print("\nInfo companies:")
companies.info()

# =========================
# 2. Xử lý NULL trong stocks1
# =========================
stocks1["high"] = stocks1.groupby("symbol")["high"]\
    .transform(lambda x: x.fillna(x.max()))

stocks1["low"] = stocks1.groupby("symbol")["low"]\
    .transform(lambda x: x.fillna(x.min()))

# =========================
# 3. Gộp stocks1 & stocks2
# =========================
stocks = pd.concat([stocks1, stocks2], ignore_index=True)
print("\n3. 15 dòng cuối stocks:")
print(stocks.tail(15))

# =========================
# 4. Gộp stocks với companies
# =========================
stocks_companies = pd.merge(stocks, companies, on="symbol", how="left")
print("\n4. stocks_companies head:")
print(stocks_companies.head())

# =========================
# 5. Giá & volume trung bình mỗi công ty
# =========================
print("\n5. Mean open, high, low, close, volume:")
print(
    stocks_companies.groupby("symbol")[["open","high","low","close","volume"]]
    .mean()
)

# =========================
# 6. Close trung bình, max, min
# =========================
print("\n6. Close mean, max, min:")
print(
    stocks_companies.groupby("symbol")["close"]
    .agg(["mean","max","min"])
)

# =========================
# 7. parsed_time
# =========================
stocks_companies["parsed_time"] = pd.to_datetime(stocks_companies["date"])
print("\n7. parsed_time dtype:")
print(stocks_companies["parsed_time"].dtype)
print(stocks_companies[["parsed_time"]].head())

# =========================
# 8. result
# =========================
stocks_companies["result"] = np.where(
    stocks_companies["close"] > stocks_companies["open"],
    "up",
    "down"
)

print("\n8. result:")
print(stocks_companies[["open","close","result"]].head())
```

```python
import pandas as pd
import numpy as np
from io import StringIO

# =========================
# DỮ LIỆU movies.csv
# =========================
movies_csv = """movieId,title,genres
1,Toy Story (1995),Adventure|Animation|Children|Comedy|Fantasy
2,Jumanji (1995),Adventure|Children|Fantasy
3,Grumpier Old Men (1995),Comedy|Romance
4,Waiting to Exhale (1995),Comedy|Drama
"""

movies = pd.read_csv(StringIO(movies_csv))

# =========================
# DỮ LIỆU ratings.csv
# =========================
ratings_csv = """userId,movieId,rating,timestamp
1,1,4.0,964982703
1,3,4.0,964981247
2,1,5.0,964982224
2,2,3.0,964983815
3,1,4.5,964982931
"""

ratings = pd.read_csv(StringIO(ratings_csv))

# =========================
# DỮ LIỆU users.csv
# =========================
users_csv = """userId,gender,age
1,M,25
2,F,30
3,M,35
"""

users = pd.read_csv(StringIO(users_csv))

# ==================================================
# 1. Đọc dữ liệu & Data Structures
# ==================================================
print("1. Movies info:")
movies.info()
print("\nRatings info:")
ratings.info()
print("\nUsers info:")
users.info()

# ==================================================
# 2. Xử lý dữ liệu thiếu / không hợp lệ
# ==================================================
print("\n2. Kiểm tra dữ liệu thiếu:")
print(movies.isnull().sum())
print(ratings.isnull().sum())
print(users.isnull().sum())

# (Dataset mẫu không có thiếu → nếu có thì xử lý bằng fillna)

# ==================================================
# 3. Gộp DataFrame (ratings + movies)
# ==================================================
movie_ratings = pd.merge(ratings, movies, on="movieId", how="left")
print("\n3. movie_ratings:")
print(movie_ratings.head())

# ==================================================
# 4. Gộp tiếp với users
# ==================================================
full_data = pd.merge(movie_ratings, users, on="userId", how="left")
print("\n4. full_data:")
print(full_data.head())

# ==================================================
# 5. Lọc dữ liệu theo yêu cầu
# Ví dụ: rating >= 4
# ==================================================
high_rating = full_data[full_data["rating"] >= 4]
print("\n5. Rating >= 4:")
print(high_rating)

# ==================================================
# 6. Thống kê dữ liệu
# Rating trung bình mỗi phim
# ==================================================
print("\n6. Rating trung bình mỗi phim:")
print(
    full_data.groupby("title")["rating"]
    .mean()
)

# ==================================================
# 7. Thống kê số lượt rating mỗi phim
# ==================================================
print("\n7. Số lượt rating mỗi phim:")
print(
    full_data.groupby("title")["rating"]
    .count()
)

# ==================================================
# 8. Parsing Timestamp
# ==================================================
full_data["rating_time"] = pd.to_datetime(full_data["timestamp"], unit="s")

print("\n8. Kiểu dữ liệu rating_time:")
print(full_data["rating_time"].dtype)

print("\n5 dòng đầu rating_time:")
print(full_data[["rating_time"]].head())
```
