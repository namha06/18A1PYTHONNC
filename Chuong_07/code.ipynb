```python
import tkinter as tk

def validate_inputs(*args):
    """Kiểm tra nếu 3 ô đều có chữ thì hiện nút Submit"""
    if name_var.get() and id_var.get() and pass_var.get():
        btn_submit.grid(row=3, column=1, pady=20)
    else:
        btn_submit.grid_forget() # Ẩn nút nếu thiếu thông tin

root = tk.Tk()
root.title("Student Info Form")
root.geometry("350x250")

# Biến lưu trữ giá trị nhập vào
name_var = tk.StringVar()
id_var = tk.StringVar()
pass_var = tk.StringVar()

# Theo dõi sự thay đổi (write) của các biến
name_var.trace_add("write", validate_inputs)
id_var.trace_add("write", validate_inputs)
pass_var.trace_add("write", validate_inputs)

# Thiết kế giao diện bằng grid
tk.Label(root, text="Student Name:").grid(row=0, column=0, padx=10, pady=10, sticky="w")
tk.Entry(root, textvariable=name_var).grid(row=0, column=1)

tk.Label(root, text="Student ID:").grid(row=1, column=0, padx=10, pady=10, sticky="w")
tk.Entry(root, textvariable=id_var).grid(row=1, column=1)

tk.Label(root, text="Password:").grid(row=2, column=0, padx=10, pady=10, sticky="w")
tk.Entry(root, textvariable=pass_var, show="*").grid(row=2, column=1)

# Nút Submit (ban đầu không hiện)
btn_submit = tk.Button(root, text="Submit", command=lambda: print("Gửi thành công!"))

root.mainloop()
```

```python
import tkinter as tk

def reverse_string_manual():
    # Lấy dữ liệu từ ô nhập liệu
    input_text = entry_word.get()
    
    # Thuật toán đảo ngược thủ công không dùng hàm định sẵn
    reversed_text = ""
    # Duyệt từ chỉ số cuối cùng về 0
    for i in range(len(input_text) - 1, -1, -1):
        reversed_text += input_text[i]
    
    # Hiển thị kết quả lên nhãn label_result
    label_result.config(text=reversed_text)

# Khởi tạo cửa sổ chính
root = tk.Tk()
root.title("Bài 7.6: Reverse Word")
root.geometry("400x250")

# Thành phần giao diện
tk.Label(root, text="Enter a word:", font=("Arial", 12)).grid(row=0, column=0, padx=20, pady=30)

# Ô nhập liệu
entry_word = tk.Entry(root, font=("Arial", 12), width=20)
entry_word.grid(row=0, column=1, padx=10, pady=30)

# Nhãn hiển thị kết quả đảo ngược
label_result = tk.Label(root, text="", font=("Arial", 12), fg="blue")
label_result.grid(row=1, column=1, sticky="w")

# Nút bấm Validate để thực hiện đảo ngược
btn_validate = tk.Button(
    root, 
    text="Validate", 
    command=reverse_string_manual,
    font=("Arial", 10),
    width=10
)
btn_validate.grid(row=2, column=1, pady=20, sticky="e")

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def calculate_sum():
    try:
        # Lấy giá trị N từ ô nhập liệu
        n_str = entry_n.get()
        if not n_str:
            messagebox.showwarning("Cảnh báo", "Vui lòng nhập giá trị N!")
            return
            
        n = int(n_str)
        
        if n < 1:
            label_result.config(text="N phải lớn hơn hoặc bằng 1")
            return

        # Tính tổng và xây dựng chuỗi hiển thị
        total = 0
        expression = "The sum is "
        
        for i in range(1, n + 1):
            total += i
            if i == 1:
                expression += "1"
            elif i <= 3:
                expression += f" + {i}"
            elif i == n:
                # Nếu n > 3, thêm dấu ba chấm trước số cuối cùng
                if n > 4:
                    expression += " + ... "
                expression += f" + {n}"
        
        expression += f" = {total}"
        
        # Cập nhật kết quả lên màn hình
        label_result.config(text=expression)
        
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng chỉ nhập số nguyên!")

# Khởi tạo cửa sổ
root = tk.Tk()
root.title("Bài 7.7: Tính tổng dãy số")
root.geometry("450x200")

# Giao diện
frame = tk.Frame(root, padx=20, pady=20)
frame.pack(expand=True)

# Dòng nhập N
tk.Label(frame, text="Enter value of integer N :", font=("Arial", 11)).grid(row=0, column=0, sticky="w")
entry_n = tk.Entry(frame, font=("Arial", 11), width=15)
entry_n.grid(row=0, column=1, padx=10)

# Dòng hiển thị kết quả
label_result = tk.Label(frame, text="", font=("Arial", 11), fg="blue")
label_result.grid(row=1, column=1, pady=10, sticky="w")

# Nút Validate
btn_validate = tk.Button(
    frame, 
    text="Validate", 
    command=calculate_sum,
    font=("Arial", 10),
    width=15
)
btn_validate.grid(row=2, column=1, pady=10)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def find_divisors():
    try:
        n_val = entry_n.get()
        if not n_val:
            messagebox.showwarning("Cảnh báo", "Vui lòng nhập giá trị N!")
            return
            
        n = int(n_val)
        divisors = [str(i) for i in range(1, n + 1) if n % i == 0]
        
        # Hiển thị các ước số cách nhau bởi dấu cách
        result_label.config(text=" ".join(divisors))
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập một số nguyên hợp lệ!")

root = tk.Tk()
root.title("Bài 7.8: Tìm ước số")
root.geometry("400x200")

# Giao diện
tk.Label(root, text="Enter the value of N:").grid(row=0, column=0, padx=10, pady=20, sticky="w")
entry_n = tk.Entry(root)
entry_n.grid(row=0, column=1, padx=10, pady=20)

tk.Label(root, text="The divisors of N :").grid(row=1, column=0, padx=10, pady=10, sticky="w")
result_label = tk.Label(root, text="", fg="blue")
result_label.grid(row=1, column=1, padx=10, pady=10, sticky="w")

btn_validate = tk.Button(root, text="Validate", command=find_divisors, width=15)
btn_validate.grid(row=2, column=1, pady=20)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
import math

def calculate():
    try:
        m = int(entry_m.get())
        n = int(entry_n.get())
        func = combo_func.get()
        
        if func == "GCD":
            # Tính ước chung lớn nhất
            res = math.gcd(m, n)
            result_text.set(f"GCD(m , n) = {res}")
        elif func == "LCM":
            # Tính bội chung nhỏ nhất: (m*n) / GCD(m,n)
            res = abs(m * n) // math.gcd(m, n)
            result_text.set(f"LCM(m , n) = {res}")
        else:
            messagebox.showwarning("Cảnh báo", "Vui lòng chọn chức năng!")
            
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập số nguyên hợp lệ cho m và n!")

root = tk.Tk()
root.title("Bài 7.9: GCD & LCM")
root.geometry("350x250")

result_text = tk.StringVar()

# Hàng nhập m
tk.Label(root, text="Enter value of m:").grid(row=0, column=0, padx=10, pady=10, sticky="w")
entry_m = tk.Entry(root)
entry_m.grid(row=0, column=1)

# Hàng nhập n
tk.Label(root, text="Enter value of n:").grid(row=1, column=0, padx=10, pady=10, sticky="w")
entry_n = tk.Entry(root)
entry_n.grid(row=1, column=1)

# Lựa chọn chức năng
tk.Label(root, text="Choose function:").grid(row=2, column=0, padx=10, pady=10, sticky="w")
combo_func = ttk.Combobox(root, values=["GCD", "LCM"], state="readonly")
combo_func.grid(row=2, column=1)
combo_func.set("GCD") # Mặc định chọn GCD

# Hiển thị kết quả
label_res = tk.Label(root, textvariable=result_text, fg="red", font=("Arial", 10, "bold"))
label_res.grid(row=3, column=1, pady=10, sticky="w")

# Nút tính toán (Tương tự nút Validate hoặc nhấn chọn chức năng để tính)
btn_calc = tk.Button(root, text="Calculate", command=calculate, width=15)
btn_calc.grid(row=4, column=1, pady=10)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show_greeting():
    name = entry_name.get()
    age_str = entry_age.get()
    
    if not name or not age_str:
        messagebox.showwarning("Thông báo", "Vui lòng nhập đầy đủ tên và tuổi!")
        return
        
    try:
        age = int(age_str)
        greeting = f"Xin chào {name}!\n"
        
        # Kiểm tra điều kiện tuổi theo yêu cầu đề bài
        if age >= 18:
            greeting += "Bạn đã trưởng thành!"
        else:
            greeting += "Bạn còn nhỏ tuổi!"
            
        messagebox.showinfo("Kết quả", greeting)
    except ValueError:
        messagebox.showerror("Lỗi", "Tuổi phải là một số nguyên!")

root = tk.Tk()
root.title("Bài 7.10: Thông báo tuổi")
root.geometry("300x200")

tk.Label(root, text="Nhập tên:").pack(pady=5)
entry_name = tk.Entry(root)
entry_name.pack()

tk.Label(root, text="Nhập tuổi:").pack(pady=5)
entry_age = tk.Entry(root)
entry_age.pack()

tk.Button(root, text="Xác nhận", command=show_greeting).pack(pady=20)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

def convert_to_lunar():
    try:
        year = int(entry_year.get())
        
        # Thuật toán tính Can Chi
        can_list = ["Canh", "Tân", "Nhâm", "Quý", "Giáp", "Ất", "Bính", "Đinh", "Mậu", "Kỷ"]
        chi_list = ["Thân", "Dậu", "Tuất", "Hợi", "Tý", "Sửu", "Dần", "Mão", "Thìn", "Tỵ", "Ngọ", "Mùi"]
        
        can = can_list[year % 10]
        chi = chi_list[year % 12]
        lunar_year = f"{can} {chi}"
        
        messagebox.showinfo("Kết quả", f"Năm {year} là năm: {lunar_year}")
        
        # Hiển thị ảnh con giáp tương ứng (Giả sử tên file ảnh là tên Chi viết thường, ví dụ: ty.png)
        # Lưu ý: Cần có thư viện Pillow (pip install Pillow)
        try:
            img_name = f"{chi.lower()}.png"
            img = Image.open(img_name)
            img = img.resize((150, 150))
            photo = ImageTk.PhotoImage(img)
            
            label_img.config(image=photo)
            label_img.image = photo # Giữ reference
        except:
            label_img.config(text=f"(Không tìm thấy ảnh {img_name})", image="")
            
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập năm là số nguyên!")

root = tk.Tk()
root.title("Bài 7.11: Chuyển đổi năm Âm lịch")
root.geometry("400x400")

tk.Label(root, text="Nhập năm Dương lịch:", font=("Arial", 12)).pack(pady=10)
entry_year = tk.Entry(root, font=("Arial", 12))
entry_year.pack()

tk.Button(root, text="Chuyển đổi", command=convert_to_lunar, bg="lightblue").pack(pady=10)

# Khu vực hiển thị ảnh con giáp
label_img = tk.Label(root)
label_img.pack(pady=20)

root.mainloop()
```

```python
import tkinter as tk

def calculate_bmi():
    try:
        weight = float(entry_weight.get())
        height = float(entry_height.get())
        
        # Công thức tính BMI
        bmi = weight / (height ** 2)
        
        # Xác định kết luận
        if bmi < 18.5:
            conclusion = "Gầy"
        elif 18.5 <= bmi < 24.9:
            conclusion = "Bình thường"
        elif 25 <= bmi < 29.9:
            conclusion = "Tiền béo phì"
        else:
            conclusion = "Béo phì"
            
        # Cập nhật hiển thị kết quả
        label_bmi_val.config(text=f"BMI: {bmi:.2f}")
        label_conclusion.config(text=f"Kết luận: {conclusion}")
    except ValueError:
        label_bmi_val.config(text="Lỗi: Nhập sai số!")
        label_conclusion.config(text="")

root = tk.Tk()
root.title("BMI Calculator")
root.geometry("300x250")

# Giao diện
tk.Label(root, text="Cân nặng (kg):").pack(pady=5)
entry_weight = tk.Entry(root)
entry_weight.pack()

tk.Label(root, text="Chiều cao (m):").pack(pady=5)
entry_height = tk.Entry(root)
entry_height.pack()

tk.Button(root, text="Tính BMI", command=calculate_bmi).pack(pady=15)

label_bmi_val = tk.Label(root, text="", font=("Arial", 10, "bold"))
label_bmi_val.pack()

label_conclusion = tk.Label(root, text="", font=("Arial", 10, "italic"), fg="blue")
label_conclusion.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show_exercise_info(chapter, lesson):
    # Nội dung ví dụ cho bài 1.1 theo yêu cầu đề bài
    if chapter == 1 and lesson == 1:
        title = "Lập trình hướng đối tượng"
        content = "Giải phương trình bậc 2"
    else:
        title = f"Chương {chapter}"
        content = f"Nội dung bài tập {chapter}.{lesson}"
        
    messagebox.showinfo(title, content)

root = tk.Tk()
root.title("Hệ thống bài tập Python nâng cao")
root.geometry("500x350")

# Tạo thanh Menu chính
menubar = tk.Menu(root)

# Tạo các Menu con cho từng chương (từ Chương 1 đến Chương 4)
for c in range(1, 5):
    chapter_menu = tk.Menu(menubar, tearoff=0)
    # Thêm 9 bài tập cho mỗi chương
    for l in range(1, 10):
        label_name = f"{c}.{l}"
        # Sử dụng lambda với tham số mặc định để tránh lỗi tham chiếu vòng lặp
        chapter_menu.add_command(
            label=label_name, 
            command=lambda ch=c, le=l: show_exercise_info(ch, le)
        )
    
    menubar.add_cascade(label=f"Chương {c}", menu=chapter_menu)

# Thêm nút Exit
menubar.add_command(label="Exit", command=root.quit)

# Cấu hình root sử dụng menubar đã tạo
root.config(menu=menubar)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def calculate_fare():
    try:
        km = float(entry_km.get())
        wait_time = int(entry_wait.get())
        vehicle_type = var_vehicle.get()
        
        fare = 0
        
        # 1. Tính tiền chờ: 5 phút đầu miễn phí, từ phút thứ 6 là 750đ/phút
        wait_fee = 0
        if wait_time > 5:
            wait_fee = (wait_time - 5) * 750
            
        # 2. Tính tiền di chuyển theo loại xe
        if vehicle_type == 4:  # Xe 4 chỗ
            if km <= 0.8:
                fare = 12000
            elif km <= 30:
                fare = 12000 + (km - 0.8) * 15300
            else:
                fare = 12000 + (30 - 0.8) * 15300 + (km - 30) * 12100
        else:  # Xe 7 chỗ
            if km <= 0.8:
                fare = 12000
            elif km <= 30:
                fare = 12000 + (km - 0.8) * 16100
            else:
                fare = 12000 + (30 - 0.8) * 16100 + (km - 30) * 13800
                
        total_total = fare + wait_fee
        label_result.config(text=f"Tổng cước: {total_total:,.0f} VNĐ")
        
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập số km và thời gian chờ hợp lệ!")

root = tk.Tk()
root.title("Bài 7.14: Tính cước Taxi")
root.geometry("400x450")

# Giao diện chọn loại xe
tk.Label(root, text="CHƯƠNG TRÌNH TÍNH CƯỚC TAXI", font=("Arial", 12, "bold")).pack(pady=10)

var_vehicle = tk.IntVar(value=4)
frame_radio = tk.LabelFrame(root, text="Loại xe", padx=10, pady=10)
frame_radio.pack(padx=20, pady=10, fill="x")

tk.Radiobutton(frame_radio, text="Xe 4 chỗ", variable=var_vehicle, value=4).pack(side="left", padx=20)
tk.Radiobutton(frame_radio, text="Xe 7 chỗ", variable=var_vehicle, value=7).pack(side="left", padx=20)

# Giao diện nhập liệu
frame_input = tk.Frame(root)
frame_input.pack(pady=10)

tk.Label(frame_input, text="Số km di chuyển:").grid(row=0, column=0, sticky="w", pady=5)
entry_km = tk.Entry(frame_input)
entry_km.grid(row=0, column=1, padx=10)

tk.Label(frame_input, text="Thời gian chờ (phút):").grid(row=1, column=0, sticky="w", pady=5)
entry_wait = tk.Entry(frame_input)
entry_wait.insert(0, "0") # Mặc định là 0 phút
entry_wait.grid(row=1, column=1, padx=10)

# Nút tính toán và kết quả
tk.Button(root, text="Tính tiền", command=calculate_fare, bg="green", fg="white", font=("Arial", 10, "bold"), width=15).pack(pady=20)

label_result = tk.Label(root, text="Tổng cước: 0 VNĐ", font=("Arial", 12, "bold"), fg="red")
label_result.pack()

root.mainloop()
```
