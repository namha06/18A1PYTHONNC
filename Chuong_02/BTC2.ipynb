```python
from xml.etree.ElementTree import Element, SubElement, tostring
from xml.dom import minidom

# Danh s√°ch sinh vi√™n m·∫´u (b·∫°n c√≥ th·ªÉ th√™m ho·∫∑c s·ª≠a t√πy √Ω)
students = [
    {"MaSV": "SV001", "HoTen": "Nguy·ªÖn VƒÉn A", "NamSinh": "2001", "Lop": "CNTT1", "GioiTinh": "Nam"},
    {"MaSV": "SV002", "HoTen": "Tr·∫ßn Th·ªã B", "NamSinh": "2002", "Lop": "CNTT2", "GioiTinh": "N·ªØ"},
    {"MaSV": "SV003", "HoTen": "L√™ VƒÉn C", "NamSinh": "2000", "Lop": "KTPM1", "GioiTinh": "Nam"}
]

# H√†m t·∫°o XML
def tao_file_xml(ds_sv):
    root = Element("DanhSachSinhVien")

    for sv in ds_sv:
        sv_tag = SubElement(root, "SinhVien")
        for key, value in sv.items():
            child = SubElement(sv_tag, key)
            child.text = value

    # L√†m ƒë·∫πp XML
    xml_str = minidom.parseString(tostring(root, 'utf-8')).toprettyxml(indent="  ", encoding="utf-8")

    # L∆∞u file XML
    with open("students.xml", "wb") as f:
        f.write(xml_str)

    # In ra n·ªôi dung XML
    print(xml_str.decode("utf-8"))
    print("\n‚úÖ ƒê√£ t·∫°o file students.xml th√†nh c√¥ng!")

# G·ªçi h√†m t·∫°o XML
tao_file_xml(students)
```

```python
import requests
from xml.dom import minidom
import csv
url = "http://www.hindustantimes.com/rss/topnews/rssfeed.xml"
response = requests.get(url)

with open("rssfeed.xml", "wb") as f:
    f.write(response.content)

print("‚úÖ ƒê√£ t·∫£i RSS feed v√† l∆∞u th√†nh file rssfeed.xml")

# 2Ô∏è‚É£ Ph√¢n t√≠ch c√∫ ph√°p t·ªáp XML
doc = minidom.parse("rssfeed.xml")

items = doc.getElementsByTagName("item")
news_list = []

for item in items:
    title = item.getElementsByTagName("title")[0].firstChild.data if item.getElementsByTagName("title") else ""
    link = item.getElementsByTagName("link")[0].firstChild.data if item.getElementsByTagName("link") else ""
    pubDate = item.getElementsByTagName("pubDate")[0].firstChild.data if item.getElementsByTagName("pubDate") else ""
    description = item.getElementsByTagName("description")[0].firstChild.data if item.getElementsByTagName("description") else ""
    
    news_list.append({
        "title": title,
        "link": link,
        "pubDate": pubDate,
        "description": description
    })

print(f"‚úÖ ƒê√£ ƒë·ªçc ƒë∆∞·ª£c {len(news_list)} tin t·ª©c t·ª´ RSS feed.")

with open("news.csv", "w", newline="", encoding="utf-8") as csvfile:
    fieldnames = ["title", "link", "pubDate", "description"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(news_list)

print("‚úÖ ƒê√£ l∆∞u danh s√°ch tin t·ª©c v√†o file news.csv th√†nh c√¥ng!")

print("\nüì¢ 3 tin ƒë·∫ßu ti√™n:")
for n in news_list[:3]:
    print("-", n["title"])
```

```python
import requests
import json

# ƒê·ªãa ch·ªâ API
url = "https://jsonplaceholder.typicode.com/posts"

print("=== B√ÄI 2.12: ƒê·ªåC V√Ä HI·ªÇN TH·ªä C√ÅC B√ÄI POST T·ª™ API ===")

try:
    # G·ª≠i y√™u c·∫ßu GET ƒë·∫øn API
    response = requests.get(url)
    response.raise_for_status()  # Ki·ªÉm tra l·ªói HTTP

    # Chuy·ªÉn d·ªØ li·ªáu nh·∫≠n ƒë∆∞·ª£c th√†nh danh s√°ch JSON
    posts = response.json()

    # Hi·ªÉn th·ªã t·ªïng s·ªë b√†i post
    print(f"T·ªïng s·ªë b√†i post: {len(posts)}\n")

    # In ra danh s√°ch c√°c post
    for post in posts:
        print(f"UserID: {post['userId']}")
        print(f"ID: {post['id']}")
        print(f"Title: {post['title']}")
        print(f"Body: {post['body']}\n{'-'*40}\n")

    # Ghi d·ªØ li·ªáu ra file JSON ƒë·ªÉ l∆∞u l·∫°i (n·∫øu mu·ªën)
    with open("posts.json", "w", encoding="utf-8") as f:
        json.dump(posts, f, ensure_ascii=False, indent=4)

    print("‚úÖ ƒê√£ l∆∞u d·ªØ li·ªáu API v√†o file posts.json")

except requests.exceptions.RequestException as e:
    print(f"‚ùå L·ªói khi k·∫øt n·ªëi API: {e}")
```
